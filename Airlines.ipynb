Import and read File

```python
import zipfile
import pandas as pd
import numpy as np
```

```python
df_airlines = pd.read_csv('/content/Airlines/archive.zip',compression='zip')
df_airlines = df_airlines.rename( columns={"Arrival Airport":"Departure Airport Code"})
df_airlines
```

Check data types

```python
df_airlines.dtypes
```

```python
print(len(df_airlines['Passenger ID']))
print(len(df_airlines))
```

Data Shape

```python
print(df_airlines.shape)
```

Check Blanks

```python
df_airlines_full = df_airlines.dropna()
df_airlines_full.shape
```

Sort data by date ( latest to newest)

```python
df_airlines_sort = df_airlines_full.sort_values(by = 'Departure Date', ascending =True)
df_airlines_sort
```

```python
df_airlines_sort = df_airlines_sort.drop(columns= "Pilot Name" )
df_airlines_sort
```

Mean Age of Travellers

```python
mean_age=df_airlines_sort['Age'].mean()
print(mean_age.round())
```

Gender Counts

```python
gender_counts = df_airlines_sort['Gender'].value_counts()
print(gender_counts)
```

Flight Status

```python
flight_status = df_airlines_sort['Flight Status'].value_counts()
print(flight_status)
```

```python
nationality = df_airlines_sort['Nationality'].value_counts()
print(nationality)
```

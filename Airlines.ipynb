```python
import zipfile
import pandas as pd
import numpy as np
```

```python
df_airlines = pd.read_csv('/content/Airlines/archive.zip',compression='zip')
df_airlines = df_airlines.rename( columns={"Arrival Airport":"Departure Airport Code"})
df_airlines
```

```python
df_airlines.dtypes
```

```python
print(df_airlines.shape)
```

```python
df_airlines_full = df_airlines.dropna()
df_airlines_full.shape
```

```python
df_airlines_sort = df_airlines_full.sort_values(by = 'Departure Date', ascending =True)
df_airlines_sort
```

```python
df_airlines_sort = df_airlines_sort.drop(columns= "Pilot Name" )
df_airlines_sort
```

```python
mean_age=df_airlines_sort['Age'].mean()
print(mean_age.round())
```
